```python
import pandas as pd
import numpy as np
df = pd.read_csv('Fish.csv')
df
```

```python
import matplotlib.pyplot as plt
df.hist(bins =30 , figsize=(20,15))
%matplotlib inline
plt.show()
```

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder(sparse=False)
columns_to_one_hot = ['Species']
df[enc.get_feature_names_out()] = enc.fit_transform(df.loc[:,columns_to_one_hot])
df.drop(columns_to_one_hot, axis=1, inplace=True)
```

```python
x = pd.DataFrame(df.iloc[:,:-1])
y =pd.DataFrame(df.iloc[:,-1])
```

Independance variable

```python
y
```

We are going to split our data

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size =0.2 ,random_state=43)
```

we want to have a look at data

```python
print (x_train.shape)
print (x_test.shape)
print (y_train.shape)
print (y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x_train , y_train)
```

We have to look on the co-efficient between the Attributes we had

```python
v = pd.DataFrame(reg.coef_,index=['Co-efficient']).transpose()
w = pd.DataFrame(x.columns, columns = ['Attributes'])
```

```python
coff_df = pd.concat([w,v] ,axis =1 ,join = 'inner')
coff_df
```

We have to compare prediction and test 

```python
y_pred = reg.predict(x_test)
y_pred = pd.DataFrame(y_pred , columns = ['Predicted'])
y_pred
```

```python
y_test
```

```python
from sklearn import metrics
print ('mean absolute error' , metrics.mean_absolute_error(y_test,y_pred))
print ('mean squared error' , metrics.mean_squared_error(y_test,y_pred))
print ('root mean squared error',np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```
